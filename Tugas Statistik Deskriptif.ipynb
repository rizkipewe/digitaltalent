```python
jumlah = data_pinjaman.nilai_pinjaman.sum()
rata = data_pinjaman.nilai_pinjaman.mean()
maks = data_pinjaman.nilai_pinjaman.max()
minim = data_pinjaman.nilai_pinjaman.min()

print("Jumlah Nilai data pinjaman ialah", jumlah, "juta")
print("Rata-rata Nilai data pinjaman ialah", rata, "juta")
print("Maksimum Nilai data pinjaman ialah", maks, "juta")
print("Minimum Nilai data pinjaman ialah", minim, "juta")
```

# 3.Visiualisasi Data

# Grafik Nilai Peminjaman Berdasarkan Tahun

```python
datapertahun = data_pinjaman.groupby('tahun').mean()
plt.plot(datapertahun, linestyle='--')
plt.xlabel('tahun')
plt.ylabel('jumlah pinjaman')
plt.show()
```

# Grafik Nilai Peminjaman Berdasarkan Jenis Bank

```python
datajenis = datapinjaman.groupby('jenis_bank').mean()
datajenis = datajenis.plot(kind='bar', 
             title ="Rata-Rata Berdasarkan Jenis Bank", figsize = (10,5), legend=True, fontsize=10, color='lightgreen')
datajenis.set_xlabel("Jenis Bank", fontsize=10)
datajenis.set_ylabel("Jumlah", fontsize=10)
```

# Grafik Nilai Peminjaman Berdasarkan Tujuan Peminjaman

```python
datatujuan=datapinjaman.groupby('tujuan_pinjaman').sum()
print (datatujuan)
activities = ['Usaha', 'Bukan Usaha']
colors = ['fuchsia', 'blue']
plt.pie(datatujuan, labels=activities, colors=colors, startangle=90, autopct='%.1f%%')
plt.show()
```

# Grafik Nilai Peminjaman Berdasarkan Tahun 

```python
dataketerangan = datapinjaman.groupby('keterangan_tujuan_pinjaman').mean()
dataketerangan = dataketerangan.plot(kind='bar', 
             title ="Rata-Rata Berdasarkan tujuan", figsize = (10,5), legend=True, fontsize=12, color='orangered')
dataketerangan.set_xlabel("Tujuan Meminjam", fontsize=12)
dataketerangan.set_ylabel("Jumlah", fontsize=12)
```
