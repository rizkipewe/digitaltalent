```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')
#fungsi untuk query dijalankan ditampilkan
def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

Report the account representative for each customer.

```python
run("""SELECT cus.customerName,cus.customerName,employ.employeeNumber,
    CONCAT(employ.firstName,' ',employ.lastName) AS Account_Representative
FROM customers cus
LEFT JOIN employees employ 
ON cus.salesRepEmployeeNumber = employ.employeeNumber
ORDER BY employ.firstName DESC
""")
```

How many orders have been placed by Herkku Gifts?

Who are the employees in Boston?

```python
run(""" SELECT employ.employeeNumber,CONCAT(employ.firstName, employ.lastName) AS FullName
    FROM employees employ
    INNER JOIN offices offi
    ON employ.officeCode = offi.officeCode
    WHERE offi.city = 'Boston'
""")
```

Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
run(""" SELECT cus.customerNumber, cus.customerName, SUM(pay.amount) AS TotalPayment
    FROM customers cus
    INNER JOIN payments pay
    ON cus.customerNumber = pay.customerNumber
    GROUP BY cus.customerNumber
    HAVING TotalPayment > 100000
    ORDER BY TotalPayment DESC
""")
```

List the value of 'On Hold' orders.

```python
run(""" SELECT DISTINCT ord.orderNumber,ord.orderDate,ord.status,SUM(od.quantityOrdered*od.priceEach) AS OrderValue
    FROM orders ord
    INNER JOIN orderdetails od
    ON ord.orderNumber = od.orderNumber
    WHERE ord.status = 'on hold'
    GROUP BY ord.orderNumber
    ORDER BY OrderValue DESC
""")
```

List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""SELECT ord.orderNumber, pro.productName,ord.orderDate
FROM orders ord
INNER JOIN orderdetails od ON od.orderNumber = ord.orderNumber
INNER JOIN products pro ON od.productCode = pro.productCode
WHERE pro.productName = '1940 Ford Pickup Truck'
ORDER BY ord.orderDate DESC
""")
```

List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""SELECT cus.customerName,SUM(od.quantityOrdered*od.priceEach) AS Value
FROM customers cus
INNER JOIN orders ord ON ord.customerNumber = cus.customerNumber
INNER JOIN orderdetails od ON od.orderNumber = ord.orderNumber
GROUP BY ord.orderNumber
HAVING Value>'25000'
ORDER BY cus.customerName ASC
""")
```

List the names of products sold at less than 80% of the MSRP.

```python
run("""SELECT pro.productName,od.priceEach,(pro.MSRP*80/100) AS `80%MSRP`
FROM products pro
INNER JOIN orderdetails od ON od.productCode = pro.productCode
HAVING od.priceEach<`80%MSRP`
""")
```

Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
    SELECT pro.productName, orderNumber, priceEach, (priceEach-buyPrice)/buyPrice*100 AS Margin
    FROM orderdetails od
    INNER JOIN products pro
    ON od.productCode = pro.productCode
    Having Margin >= 100 
    ORDER BY  PriceEach, orderNumber
""")
```

List the products ordered on a Monday.

```python
run(""" SELECT pro.productName, ord.orderDate, DAYNAME(ord.orderDate) as Hari
    FROM orders ord
    INNER JOIN orderdetails od
    ON ord.orderNumber = od.orderNumber
    INNER JOIN products pro
    ON od.productCode = pro.productCode
    HAVING Hari = 'Monday'
""")
```

What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
    SELECT DISTINCT od.productCode, ord.status, SUM(od.quantityOrdered) as quantity
    FROM orders ord
    INNER JOIN orderdetails od
    ON ord.orderNumber = od.orderNumber
    WHERE ord.status = 'on hold'
    GROUP BY od.productCode
    ORDER BY productCode
""")
```
